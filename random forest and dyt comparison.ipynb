```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score
import os
from sklearn.metrics import recall_score


working_directory=os.getcwd()
path = working_directory + '/DATASET/iris.csv'
df = pd.read_csv(path)
# print(data.head())
print(df.info())



for i in range(0,len(df['pet_leng'])):
    if i<0.15:
        df['pet_leng'][i]=0
    else:
        df['pet_leng'][i]=1


df=df.drop(['no'],axis=1)
X = df.drop(['pet_leng'], axis=1)  
y = df['pet_leng'] 
print (df.head())

x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)


DTC = DecisionTreeClassifier()
DTC.fit(x_train, y_train)
y_predicted=DTC.predict(x_test)
print("\naccuracy on test data is: ",DTC.score(x_test, y_test)*100)
print("\nclasification  error is: ",1-DTC.score(x_test, y_test))
from sklearn import metrics
cm=metrics.confusion_matrix(y_test, y_predicted)

print("\nconfusion matrix: ",cm)

sensitivity1 = cm[0,0]/(cm[0,0]+cm[0,1])
print('\nSensitivity : ', sensitivity1 )

specificity1 = cm[1,1]/(cm[1,0]+cm[1,1])
print('Specificity : ', specificity1)

precision = precision_score(y_test, y_predicted)
print("precision: ",precision)



print("\n\n\n\n\n")
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(x_train, y_train)
y_predicted=DTC.predict(x_test)
print("\naccuracy on test data is: ",rfc.score(x_test, y_test)*100)
print("\nerror rate is: ",1-rfc.score(x_test, y_test))
from sklearn import metrics
cm=metrics.confusion_matrix(y_test, y_predicted)

print("\nconfusion matrix: ",cm)



precision = precision_score(y_test, y_predicted)
print("precision: ",precision)


recall = recall_score(y_test, y_predicted)
print("recall: ",recall)





```
